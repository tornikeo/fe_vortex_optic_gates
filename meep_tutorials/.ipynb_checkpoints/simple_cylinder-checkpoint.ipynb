```python
def eps_matrix(phi, coeff):
    e1 = 1 - coeff * np.sin(phi) ** 2
    e2 = coeff * np.sin(phi) * np.cos(phi)
    e3 = 1 - coeff * np.cos(phi) ** 2
    return e1,e2,e3

import matplotlib.pyplot as plt
```

```python

f,ax = plt.subplots(3,1)

phis = np.linspace(0, 2 * np.pi, 50)
coeffs = np.linspace(1,1, 50)
eps_list = eps_matrix(phis, coeffs)
phis /= np.pi

ax[0].plot(phis, eps_list[0])
ax[0].plot(phis, eps_list[1])
ax[0].plot(phis, eps_list[2])

phis = np.linspace(0, 2 * np.pi, 50)
coeffs = np.linspace(2,2, 50)
eps_list = eps_matrix(phis, coeffs)
phis /= np.pi

ax[1].plot(phis, eps_list[0])
ax[1].plot(phis, eps_list[1])
ax[1].plot(phis, eps_list[2])


phis = np.linspace(0, 2 * np.pi, 50)
coeffs = np.linspace(12,12, 50)
eps_list = eps_matrix(phis, coeffs)
phis /= np.pi

ax[2].plot(phis, eps_list[0])
ax[2].plot(phis, eps_list[1])
ax[2].plot(phis, eps_list[2])
```

```python
import numpy as np
def cart2pol(x, y):
    rho = np.sqrt(x**2 + y**2)
    phi = np.arctan2(y, x)
    return(rho, phi)

def pol2cart(rho, phi):
    x = rho * np.cos(phi)
    y = rho * np.sin(phi)
    return(x, y)
```

```python
def vortex_material(pos: mp.Vector3) -> mp.Medium:
    rho, phi = cart2pol(pos.x, pos.y)
    e1,e2,e3 = eps_matrix(phi, -20)
    return mp.Medium(epsilon_diag=mp.Vector3(e1,e3,1), 
                     epsilon_offdiag=mp.Vector3(e2,0,0))
```

```python
sx, sy = 5, 5
cell = mp.Vector3(sx, sy, 0)
# def vortex_material(pos: mp.Vector3) -> mp.Medium:
#     # if pos.x < 0:
#     #     return mp.Medium(epsilon=12)
#     # else:
#     #     return mp.Medium(epsilon=6)
#     theta = np.

geometry = [
    mp.Cylinder(
        radius=1,
        center=mp.Vector3(),
        # material= materials.Ag
        material=vortex_material
    )
]
# fcen = 1/wavelength
# df = fcen * .1

# fcen = 0.15
fcen = 1
fwidth = fcen * 0.5
source_width = 1
sources = [
    mp.Source(
        mp.GaussianSource(fcen, cutoff=5, fwidth=fwidth),
        # mp.ContinuousSource(fcen, cutoff=5, fwidth=fwidth),
        component=mp.Hz,
        center=mp.Vector3(-0.5 * sx + dpml, 0, 0),
        size=mp.Vector3(0,source_width,0)
    )
]

dpml = 1.0
pml_layers = [mp.PML(dpml)]

resolution = 35

sim = mp.Simulation(
    cell_size=cell,
    boundary_layers=pml_layers,
    geometry=geometry,
    sources=sources,
    resolution=resolution,
)

import matplotlib as mpl 

sim.plot2D(
    eps_parameters=dict(
        # contour=dict(linewidth=1),
        colorbar=dict(cmap=mpl.colormaps["plasma"], )
    ),
    colorbar_parameters=dict(pad=.5, size=.5, position="right")
)
```
